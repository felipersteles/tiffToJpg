```python
import cv2, os
import tifffile as tff

# Tem que botar a barra no final
input_path = '/content/drive/MyDrive/Polyp/DATASET/PolypDetection/test/tiff/' 
output_path = '/content/drive/MyDrive/Polyp/DATASET/PolypDetection/test/1/'
```

```python
count = 0
for infile in os.listdir(input_path):
    #print ("file : " + infile)
    read = tff.imread(input_path + infile)
    outfile = infile.split('.')[0] + '.jpg'
    tff.imwrite(output_path + outfile, read, [int(cv2.IMWRITE_JPEG_QUALITY), 200])
    count += 1
```

```python
print("qtd_input:", count)

print( "qtd_output:" , len(os.listdir(output_path)))
```
